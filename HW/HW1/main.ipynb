# II. Cài đặt các hàm cần thiết

```python
# Hàm tạo ma trận vuông ngẫu nhiên với kích thước n
def generate_matrix(n):
    return [[random.randint(0, 9) for _ in range(n)] for _ in range(n)]

# Hàm nhân 2 ma trận 2 chiều lại với nhau
def multiply_matrices(matrix1, matrix2):
    n = len(matrix1)
    result = [[0] * n for _ in range(n)]
    for i in range(n):
        for j in range(n):
            for k in range(n):
                result[i][j] += matrix1[i][k] * matrix2[k][j]
    return result

# In ma trận
def print_matrix(matrix):
    for row in matrix:
        print(" ".join(map(str, row)))
```

## 1. Vẽ biểu đồ

### Import thư viện và tạo DataFrame

```python
import pandas as pd
import matplotlib.pyplot as plt

# Dữ liệu
data = {
    "Phương pháp": ["C", "C", "C", "C", "C", "C", "Python", "Python", "Python", "Python", "Python", "Python",
                   "Numpy", "Numpy", "Numpy", "Numpy", "Numpy", "Numpy"],
    "Size": [5, 10, 20, 50, 100, 1000, 5, 10, 20, 50, 100, 1000, 5, 10, 20, 50, 100, 1000],
    "CPU times (ms)": [12.1, 10.4, 11.4, 11.4, 8.3, 74.7, 0.058, 0.216, 0.673, 9.25, 62.9, 122000, 
                       0.342, 0.757, 1.17, 3.33, 3.42, 709],
    "Wall time (ms)": [108, 107, 108, 109, 113, 2888, 0.059, 0.218, 0.676, 9.01, 61.2, 122000, 
                       0.252, 0.496, 0.606, 1.5, 1.71, 707]
}

# Tạo DataFrame
df = pd.DataFrame(data)

```

### a. So sánh <code>CPU times</code> và <code>Wall time</code> cho từng phương pháp

```python
# Vẽ biểu đồ cho phương pháp C
plt.figure(figsize=(12, 5))

method = 'C'
method_data = df[df['Phương pháp'] == method]
plt.plot(method_data['Size'], method_data['CPU times (ms)'], marker='o', color='blue', label='CPU times')
plt.plot(method_data['Size'], method_data['Wall time (ms)'], marker='x', color='orange', label='Wall time')
plt.title(f'Phương pháp {method}')
plt.xlabel('Kích thước (Size)')
plt.ylabel('Thời gian (ms)')
plt.legend()
plt.ylim(0, max(method_data['CPU times (ms)'].max(), method_data['Wall time (ms)'].max()) + 100)

plt.tight_layout()
plt.show()

# Vẽ biểu đồ cho phương pháp Numpy
plt.figure(figsize=(12, 5))

method = 'Numpy'
method_data = df[df['Phương pháp'] == method]
plt.plot(method_data['Size'], method_data['CPU times (ms)'], marker='o', color='blue', label='CPU times')
plt.plot(method_data['Size'], method_data['Wall time (ms)'], marker='x', color='orange', label='Wall time')
plt.title(f'Phương pháp {method}')
plt.xlabel('Kích thước (Size)')
plt.ylabel('Thời gian (ms)')
plt.legend()
plt.ylim(0, max(method_data['CPU times (ms)'].max(), method_data['Wall time (ms)'].max()) + 100)

plt.tight_layout()
plt.show()

# Vẽ biểu đồ cho phương pháp Python
plt.figure(figsize=(12, 5))

method = 'Python'
method_data = df[df['Phương pháp'] == method]
plt.plot(method_data['Size'], method_data['CPU times (ms)'], marker='o', color='blue', label='CPU times')
plt.plot(method_data['Size'], method_data['Wall time (ms)'], marker='x', color='orange', label='Wall time')
plt.title(f'Phương pháp {method}')
plt.xlabel('Kích thước (Size)')
plt.ylabel('Thời gian (ms)')
plt.legend()
plt.ylim(0, max(method_data['CPU times (ms)'].max(), method_data['Wall time (ms)'].max()) + 100)

plt.tight_layout()
plt.show()
```

### b. Vẽ biểu đồ <code>CPU times</code>

```python
# Vẽ biểu đồ CPU times
plt.figure(figsize=(12, 6))
for method in df["Phương pháp"].unique():
    subset = df[df["Phương pháp"] == method]
    plt.plot(subset["Size"], subset["CPU times (ms)"], marker='o', label=f'{method} - CPU times')

plt.yscale('log')  # Sử dụng thang đo logarit cho trục y
plt.xlabel('Kích thước ma trận (n)')
plt.ylabel('Thời gian CPU (ms)')
plt.title('So sánh thời gian CPU cho các phương pháp')
plt.legend()
plt.grid(True)
plt.xticks(subset["Size"])  # Đặt các giá trị kích thước ma trận
plt.tight_layout()
plt.show()
```

### c. Vẽ biểu đồ <code>Wall time</code>

```python
# Vẽ biểu đồ Wall time
plt.figure(figsize=(12, 6))
for method in df["Phương pháp"].unique():
    subset = df[df["Phương pháp"] == method]
    plt.plot(subset["Size"], subset["Wall time (ms)"], marker='x', linestyle='--', label=f'{method} - Wall time')

plt.yscale('log')  # Sử dụng thang đo logarit cho trục y
plt.xlabel('Kích thước ma trận (n)')
plt.ylabel('Thời gian Wall (ms)')
plt.title('So sánh thời gian Wall cho các phương pháp')
plt.legend()
plt.grid(True)
plt.xticks(subset["Size"])  # Đặt các giá trị kích thước ma trận
plt.tight_layout()
plt.show()
```

### 2.1. Nhận xét <code>CPU times</code> và Wall time
#### a. Chương trình C:
**Nhận xét:**
- Wall time có sự chênh lệch lớn (cao hơn) so với <code>CPU times</code> ở mọi trường hợp.
- Wall time có sự chênh lệch càng lớn khi tăng dần kích thước lên, và cao nhất ở kích thước 1000
  
**Lý giải:**
- <code>Wall time</code> là thời gian thực tế người dùng phải đợi để chương trình có thể xử lý xong tác vụ nó bao gồm thời gian CPU xử lý, hệ thống, I/O và nhiều vấn đề khác nữa do đó nó thường sẽ lớn hơn so với <code>CPU times</code>.
- Chương trình C trong ví dụ thực hiện lập trình ở dạng đơn thuần, chỉ dùng một nhân do đó cũng góp phần là lý do khiến <code>Wall time</code> cao hơn <code>CPU times</code>.
- Chương trình C trong ví dụ sử dụng cấp phát động để cấp phát cho mảng hai chiều và giải phóng bộ nhớ cuối chương trình, do đó khi ở kích thước càng lớn thì việc giải phóng và cấp phát sẽ càng lâu (độ phức tạp O(n^2)), do đó khiến cho chương trình tốn rất nhiều thời gian để thực thi, trong khi hiệu xuất CPU xử lý rất nhanh chóng.
